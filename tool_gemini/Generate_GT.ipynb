```python
import sys
from google import genai
from google.genai import types

def interactive_hate_speech_analyzer():
    # Khởi tạo API (ĐIỀN KEY MỚI CỦA BẠN VÀO ĐÂY NHÉ, NHỚ XÓA KEY CŨ)
    client = genai.Client(api_key="")

    system_instruction = (
        "Đóng vai một chuyên gia ngôn ngữ học. Hãy đọc câu bình luận mạng xã hội sau "
        "và giải thích tại sao nó lại là ngôn từ thù địch hoặc xúc phạm.\n"
        "Quy tắc bắt buộc:\n"
        "1. Trả lời cực kỳ ngắn gọn, tuyệt đối KHÔNG VƯỢT QUÁ 30 từ.\n"
        "2. Bắt buộc trích xuất từ khóa chửi bới / thù địch trong câu gốc và bỏ vào dấu ngoặc kép ('...').\n"
        "3. Trực diện, đi thẳng vào lý do, tuyệt đối không lặp lại câu hỏi, không thêm lời dẫn."
    )

    config = types.GenerateContentConfig(
        system_instruction=system_instruction,
        temperature=0.0,
    )

    print("==================================================")
    print("🤖 CÔNG CỤ TẠO DỮ LIỆU QA BẰNG GEMINI (CHẾ ĐỘ THỦ CÔNG)")
    print("Nhập 'q' hoặc 'thoat' để dừng chương trình.")
    print("==================================================\n")

    # Vòng lặp vô hạn, cứ nhập là sinh câu trả lời
    while True:
        # Máy dừng lại chờ bạn dán câu bình luận vào
        user_input = input("👉 Dán câu bình luận HATE vào đây: ").strip()

        # Nhập chữ 'q' để thoát vòng lặp
        if user_input.lower() in ['q', 'thoat', 'exit', 'quit']:
            print("👋 Đã thoát chương trình. Chúc bạn làm data vui vẻ!")
            break

        if not user_input:
            continue

        print("⏳ Đang phân tích...")
        prompt = f"Nội dung: '{user_input}'"

        try:
            response = client.models.generate_content(
                model='gemini-2.5-flash',
                contents=prompt,
                config=config
            )
            # In ra đáp án để bạn copy dán vào Excel
            print(f"🎯 Đáp án GT: {response.text.strip()}\n")
            print("-" * 50)

        except Exception as e:
            print(f"❌ Lỗi API (Có thể do mạng lag hoặc hết hạn mức Token): {e}\n")

if __name__ == "__main__":
    # Chỉ việc chạy hàm này là màn hình terminal sẽ hiện lên chờ bạn
    interactive_hate_speech_analyzer()
```
